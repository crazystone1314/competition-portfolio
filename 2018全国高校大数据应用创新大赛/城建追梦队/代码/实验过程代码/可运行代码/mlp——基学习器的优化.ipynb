```python
def bincount2D_vectorized(a):    
    N = a.max()+1
    a_offs = a + np.arange(a.shape[0])[:,None]*N
    return np.bincount(a_offs.ravel(), minlength=a.shape[0]*N).reshape(-1,N)

S_training = training.iloc[:,[0,2,4,6,8]].astype(int)
S_training = pd.DataFrame(bincount2D_vectorized(S_training.values),columns=['suitCount0','suitCount1','suitCount2','suitCount3','suitCount4'])
training = pd.merge(training, S_training, how='left', left_index=True, right_index=True).drop(['suitCount0'],axis=1)


R_training = training.iloc[:,np.arange(1,10,2)].astype(int)
cols = ['rank{}'.format(x) for x in range(0,14,1)]
R_training = pd.DataFrame(bincount2D_vectorized(R_training.values),columns=cols)
training = pd.merge(training, R_training, how='left', left_index=True, right_index=True).drop(['rank0'],axis=1)

training.head()
```

```python
X = training.drop(['hand'],axis=1)
y = training.hand
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=24)
X_resampled, y_resampled = ros.fit_sample(X, y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.33, random_state=34)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neural_network import MLPClassifier


mlp = MLPClassifier(hidden_layer_sizes=(80, 80, 80), solver='adam', activation='tanh', alpha = 0.0001, learning_rate_init=0.001, max_iter=300)
scores = cross_val_score(mlp, X_resampled, y_resampled, scoring='accuracy')
average_accuracy = np.mean(scores)*100
print('The average accuracy is {0:.10f}%'.format(average_accuracy))
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import log_loss, roc_auc_score, auc, roc_curve, classification_report, confusion_matrix, accuracy_score
#from sklearn.model_selection import KFold, train_test_split, StratifiedKFold, cross_val_score, GridSearchCV
mlp = MLPClassifier(hidden_layer_sizes=(60, 60, 50), solver='adam', activation='relu', alpha = 0.0001, learning_rate_init=0.001, max_iter=300)

mlp.fit(X_train, y_train)

y_pred = mlp.predict(X_test)

score = accuracy_score(y_test,y_pred)
print("CV accuracy: {}".format(score))

preds_probs = mlp.predict_proba(X_test)
loss = log_loss(y_test, preds_probs)
print('loss: ', loss)
print(confusion_matrix(y_test,y_pred))  
print(classification_report(y_test,y_pred))  
```
